### Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
```

```python
lung_cancer_data = pd.read_csv("lung_cancer_prediction_dataset.csv")
air_pol = pd.read_csv("global_air_pollution_dataset.csv")
```

### Part 3. Use the datasets provided to create graphs

#### 1. Create a Boxplot of Lung Cancer Deaths Distribution. (Python & R)

```python
plt.figure(figsize=(10,6))
sns.boxplot(y=lung_cancer_data['Annual_Lung_Cancer_Deaths'])
plt.title('Boxplot of Lung Cancer Deaths Distribution')
plt.ylabel('Lung Cancer Deaths')
plt.show()
```

#### 2. Create a Histogram of PM2.5 AQI Values. (Python & R)

```python
plt.figure(figsize=(10,6))
sns.histplot(air_pol['PM2.5_AQI_Value'], kde=False, bins=50, color='grey')
plt.title('Histogram of PM2.5 AQI Values')
plt.xlabel('PM2.5 AQI')
plt.ylabel('Frequency')
plt.show()
```

#### 3. Create a Density Plot of the Lung Cancer Mortality Rate. (Python & R)

```python
plt.figure(figsize=(10,6))
sns.kdeplot(lung_cancer_data['Mortality_Rate'], fill=True, color='darkred')
plt.title('Density Plot of Lung Cancer Mortality Rate')
plt.xlabel('Lung Cancer Mortality Rate')
plt.ylabel('Density')
plt.show()
```

### Part 4. Recreate the following graphs

#### 1. Use the matplotlib library for this graph.

```python
pm25_values = air_pol['PM2.5_AQI_Value'].dropna()

plt.figure(figsize=(10, 6))

ax = sns.histplot(pm25_values, bins=30, kde=False, color='#8C8CFF', edgecolor='#6B6BFF', stat="density", zorder=2)
kde = sns.kdeplot(pm25_values, color='#990000', linewidth=2, zorder=3)

x_vals = kde.lines[0].get_xdata()
y_vals = kde.lines[0].get_ydata()

ax.set_facecolor((234/255, 234/255, 242/255))
ax.grid(True, color='white', linestyle='-', linewidth=0.3, zorder=1)

plt.fill_between(x_vals, y_vals, color='darkred', alpha=0.3)

plt.xlabel('PM2.5 AQI Value', fontweight = "bold")
plt.ylabel('Density', fontweight = "bold")
plt.title('PM2.5 AQI Distribution with Density Overlay', fontweight = "bold")
plt.suptitle('This plot represents the distribution of PM2.5 AQI values with a density overlay.', x=0.1, y=-0.05, ha='left', fontsize = 7)

hist_patch = mpatches.Patch(facecolor='#8C8CFF', edgecolor='#6B6BFF', alpha=0.6, label='Histogram')
dens_patch = mpatches.Patch(facecolor='#E0ACAE', edgecolor='#990000', linewidth=2, alpha=0.6, label='Density Plot (KDE)')

plt.legend(handles=[hist_patch, dens_patch], loc='upper right')

plt.gcf().set_facecolor('#F4F4F4')
plt.gcf().set_edgecolor('black')  
plt.gcf().set_linewidth(0.1)

plt.show()
```
